```python
import os
def fol(main_folder):
  sn1 = 'images'
  sn2= 'labels'
  sp1 = os.path.join(main_folder, sn1)
  sp2 = os.path.join(main_folder, sn2)
  os.makedirs(sp1, exist_ok=True)
  os.makedirs(sp2, exist_ok=True)

main1 = '/content/sample_data/dataset/test'
main2 ='/content/sample_data/dataset/train'
fol(main1)
fol(main2)
```

```python
imgs =[]
xmls =[]

crsPath = '/content/sample_data/Data/' #dir where images and annotations stored

#setup ratio (val ratio = rest of the files in origin dir after splitting into train and test)
train_ratio = 0.8
val_ratio = 0.2


#total count of imgs
totalImgCount = len(os.listdir(crsPath))/2

#soring files to corresponding arrays
for (dirname, dirs, files) in os.walk(crsPath):
    for filename in files:
        if filename.endswith('.txt'):
            xmls.append(filename)
        else:
            imgs.append(filename)


#counting range for cycles
countForTrain = int(len(imgs)*train_ratio)
countForVal = int(len(imgs)*val_ratio)
print("training images are : ",countForTrain)
print("Validation images are : ",countForVal)
```

```python
trainimagePath ='/content/sample_data/dataset/train/images'
trainlabelPath = '/content/sample_data/dataset/train/labels'
valimagePath = '/content/sample_data/dataset/test/images'
vallabelPath = '/content/sample_data/dataset/test/labels'
for x in range(countForTrain):

    fileJpg = choice(imgs)
    fileXml = fileJpg[:-4] +'.txt'

    shutil.copy(os.path.join(crsPath, fileJpg), os.path.join(trainimagePath, fileJpg))
    shutil.copy(os.path.join(crsPath, fileXml), os.path.join(trainlabelPath, fileXml))

    imgs.remove(fileJpg)
    xmls.remove(fileXml)

for x in range(countForVal):

    fileJpg = choice(imgs)
    fileXml = fileJpg[:-4] +'.txt'

    #move both files into train dir
    shutil.copy(os.path.join(crsPath, fileJpg), os.path.join(valimagePath, fileJpg))
    shutil.copy(os.path.join(crsPath, fileXml), os.path.join(vallabelPath, fileXml))

    #remove files from arrays
    imgs.remove(fileJpg)
    xmls.remove(fileXml)
```

```python
def calculate_yellow_pixel_percentage(image, x1, y1, x2, y2, lower_yellow, upper_yellow):
    # Extract the region of interest (ROI) from the image
    roi = image[int(y1):int(y2), int(x1):int(x2)]

    # Convert the ROI to the HSV color space
    hsv_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Threshold the HSV image to get only yellow pixels
    yellow_mask = cv2.inRange(hsv_roi, lower_yellow, upper_yellow)

    # Count the number of yellow pixels
    yellow_pixel_count = cv2.countNonZero(yellow_mask)

    # Calculate the total number of pixels in the ROI
    total_pixel_count = roi.shape[0] * roi.shape[1]

    # Calculate the percentage of yellow pixels
    yellow_pixel_percentage = (yellow_pixel_count / total_pixel_count) * 100

    return yellow_pixel_percentage
```

```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

# Define the folder path containing the images
folder_path = '/content/yolov5/runs/detect/exp/'

# Define the lower and upper bounds for yellow color in HSV
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])



# Initialize the count
count = 0

# Iterate over the files in the folder
for filename in os.listdir(folder_path):
    a=[]
    print("Processing image:", filename)
    image = cv2.imread(os.path.join(folder_path, filename))

    for i in image_bounding_boxes:
        if(i[0]==filename):
            a.append(i[1])

    for i in a:
        a,b,c,d=map(int,i)
        x1=a
        y1=b
        x2=c
        y2=d
        percentage = calculate_yellow_pixel_percentage(image, x1, y1, x2, y2, lower_yellow, upper_yellow)
        if percentage > 45:
            count += 1

print("Total count:", count)
```
